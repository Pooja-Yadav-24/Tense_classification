```python
%%writefile app.py
import streamlit as st
import pandas as pd
from simpletransformers.classification import ClassificationModel

# Load your trained model
model = ClassificationModel('bert', 'bert-base-cased', num_labels=3, use_cuda=False)
model.load_model('path_to_your_model_directory')

# Define label mappings
label_mapping_reverse = {0: 'Present', 1: 'Past', 2: 'Future'}

# Streamlit app
st.title("Tense Classification App")
st.write("Enter a sentence to classify its tense:")

# Input text box
user_input = st.text_input("Sentence")

if st.button("Classify"):
    if user_input:
        # Predict the tense
        predictions, _ = model.predict([user_input])
        predicted_label = label_mapping_reverse[predictions[0]]
        st.write(f"The predicted tense is: {predicted_label}")
    else:
        st.write("Please enter a sentence.")
```
